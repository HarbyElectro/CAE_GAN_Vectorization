# Latent Space Visualization

This notebook visualizes latent vectors exported by `CAE_CelebA.py` or `Gan_autoencoder_Cfp.py` using PCA and t-SNE.

## Setup

```python
import os
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

plt.rcParams["figure.figsize"] = (8, 6)
```

## Load Latent Vectors

```python
# Update this path based on your experiment output.
LATENTS_PATH = Path("../runs/gan_ae_cfpd/latents.npy")

if not LATENTS_PATH.exists():
    raise FileNotFoundError(f"Latent file not found: {LATENTS_PATH}")

latents = np.load(LATENTS_PATH)
print("Original latent shape:", latents.shape)

# Flatten latent tensors if needed.
# Example: (N, C, H, W) -> (N, C*H*W)
latents_flat = latents.reshape(latents.shape[0], -1)
print("Flattened latent shape:", latents_flat.shape)
```

## PCA Visualization

```python
pca = PCA(n_components=2, random_state=42)
latents_pca = pca.fit_transform(latents_flat)

plt.figure()
plt.scatter(latents_pca[:, 0], latents_pca[:, 1], s=10, alpha=0.7)
plt.title("PCA Visualization of Latent Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

print("Explained variance ratio:", pca.explained_variance_ratio_)
```

## t-SNE Visualization

```python
# For large datasets, use a subset to keep t-SNE fast.
max_points = 2000
subset = latents_flat[:max_points]

tsne = TSNE(n_components=2, random_state=42, init="pca", learning_rate="auto", perplexity=30)
latents_tsne = tsne.fit_transform(subset)

plt.figure()
plt.scatter(latents_tsne[:, 0], latents_tsne[:, 1], s=10, alpha=0.7)
plt.title("t-SNE Visualization of Latent Space")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.grid(True)
plt.show()
```

## Save PCA Coordinates

```python
OUTPUT_PATH = Path("../runs/latent_pca_coordinates.npy")
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
np.save(OUTPUT_PATH, latents_pca)
print(f"Saved PCA coordinates to: {OUTPUT_PATH}")
```
